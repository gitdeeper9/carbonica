# 2. Revelle Factor Analysis
## Ocean Buffer Capacity Decline 1960-2025

This notebook reproduces Figure 2 from the CARBONICA paper: Revelle Factor time series showing accelerating ocean buffer capacity decline.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica import CARBONICA
from carbonica.modules.ocean_sink import OceanSinkModel
from carbonica.visualization.parameter_plots import ParameterPlots

# Initialize
carbonica = CARBONICA(data_dir="../data")
ocean = OceanSinkModel()
plots = ParameterPlots(width=80)
```

```python
# Get Revelle Factor time series
years = [1750, 1800, 1850, 1900, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2025]
revelle_values = []
beta_values = []

for year in years:
    R = ocean.get_revelle_factor(year)
    beta = 1.0 / R
    revelle_values.append(R)
    beta_values.append(beta)
    print(f"{year}: R = {R:.2f}, β = {beta:.4f}")
```

```python
# Calculate acceleration
from carbonica.statistics.simple_stats import SimpleStats

# 1960-1990 period
R_1960 = ocean.get_revelle_factor(1960)
R_1990 = ocean.get_revelle_factor(1990)
rate_1960_1990 = (R_1990 - R_1960) / 30

# 2010-2025 period
R_2010 = ocean.get_revelle_factor(2010)
R_2025 = ocean.get_revelle_factor(2025)
rate_2010_2025 = (R_2025 - R_2010) / 15

print(f"\n=== Revelle Factor Acceleration ===")
print(f"1960-1990 rate: {rate_1960_1990:.3f}/year")
print(f"2010-2025 rate: {rate_2010_2025:.3f}/year")
print(f"Acceleration factor: {rate_2010_2025/rate_1960_1990:.1f}x (paper: 2.5x)")
```

```python
# Plot Revelle Factor
print(plots.plot_timeseries(years, revelle_values, "Revelle Factor 1750-2025", "R"))
```

```python
# Plot Buffer Capacity (β)
print(plots.plot_timeseries(years, beta_values, "Buffer Capacity β = 1/R 1750-2025", "β"))
```

```python
# Project future
print("\n=== Future Projections ===")
projections = ocean.project_revelle(end_year=2050)
for year, R in projections:
    print(f"{year}: R = {R:.2f}")
```

```python
# Paper validation
print("\n=== Paper Validation ===")
print(f"Pre-industrial R (1750): {ocean.get_revelle_factor(1750):.1f} (paper: 9.1)")
print(f"1960 R: {ocean.get_revelle_factor(1960):.1f} (paper: 10.2)")
print(f"2025 R: {ocean.get_revelle_factor(2025):.1f} (paper: 12.4)")
print(f"Buffer loss: {(1 - 1/12.4*9.1)*100:.0f}% (paper: 36%)")
```

```python
# Ocean uptake efficiency
print("\n=== Ocean Uptake Efficiency ===")
for year in [2025, 2030, 2040, 2050, 2060]:
    eff = ocean.get_ocean_uptake_efficiency(year)
    print(f"{year}: {eff*100:.0f}% of emissions absorbed")
```
