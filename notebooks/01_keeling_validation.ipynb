# 1. Keeling Curve Validation
## PCSI vs Atmospheric CO₂ 1960-2025

This notebook reproduces Figure 1 from the CARBONICA paper: PCSI validation against the 65-year Keeling Curve record.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica import CARBONICA
from carbonica.visualization.parameter_plots import ParameterPlots

# Initialize
carbonica = CARBONICA(data_dir="../data")
plots = ParameterPlots(width=80)
```

```python
# Load Keeling data
keeling_data = carbonica.budget.load_keeling()
print(f"Loaded {len(keeling_data)} years of Keeling data")
```

```python
# Compute PCSI for each decade
years = [1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020, 2025]
pcsi_values = []
co2_values = []

for year in years:
    pcsi = carbonica.compute_pcsi(year)
    pcsi_values.append(pcsi)
    co2 = carbonica.budget.keeling_ref.get(year, 0)
    co2_values.append(co2)
    print(f"{year}: CO₂ = {co2:.1f} ppm, PCSI = {pcsi:.3f}")
```

```python
# Calculate correlation
from carbonica.statistics.simple_stats import SimpleStats

correlation = SimpleStats.correlation(pcsi_values, co2_values)
print(f"\nCorrelation between PCSI and CO₂: {correlation:.3f}")
print(f"R² = {correlation**2:.3f} (paper: 0.947)")
```

```python
# Plot time series
print(plots.plot_timeseries(years, pcsi_values, "PCSI Time Series 1960-2025", "PCSI"))
```

```python
# Plot CO₂
print(plots.plot_timeseries(years, co2_values, "Keeling Curve CO₂ 1960-2025", "CO₂ (ppm)"))
```

```python
# Calculate growth rates
growth_rates = []
for i in range(1, len(years)):
    rate = (co2_values[i] - co2_values[i-1]) / (years[i] - years[i-1])
    growth_rates.append(rate)
    print(f"{years[i-1]}-{years[i]}: {rate:.2f} ppm/yr")
```

```python
# Paper validation
print("\n=== Paper Validation ===")
print(f"PCSI 1960: {carbonica.compute_pcsi(1960):.3f} (paper: 0.31)")
print(f"PCSI 2025: {carbonica.compute_pcsi(2025):.3f} (paper: 0.78)")
print(f"Correlation with Keeling: {correlation**2:.3f} (paper: 0.947)")
```
