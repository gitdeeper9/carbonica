# 8. SSP Scenario Projections
## PCSI Future Trajectories 2025-2070

This notebook reproduces Figure 8 from the CARBONICA paper: PCSI projections under different SSP scenarios.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica import CARBONICA
from carbonica.visualization.projection_plot import ProjectionPlot

# Initialize
carbonica = CARBONICA(data_dir="../data")
proj = ProjectionPlot()
```

```python
# Display all projections
print(proj.render_all_projections())
```

```python
# Threshold crossing years
print("\n=== Threshold Crossing Years ===")
crossings = proj.get_threshold_crossing()
for scenario, year in crossings.items():
    print(f"{scenario:10s}: PCSI ≥ 0.80 in {year}")
```

```python
# Compare with paper values
print("\n=== Comparison with Paper ===")
paper_values = {
    'SSP1-1.9': (2055, 2067),
    'SSP3-7.0': (2047, 2053),
    'SSP5-8.5': (2041, 2046)
}

for scenario, (low, high) in paper_values.items():
    crossing = crossings[scenario]
    if low <= crossing <= high:
        status = "✅"
    else:
        status = "❌"
    print(f"{scenario:10s}: {crossing} (paper: {low}-{high}) {status}")
```

```python
# Uncertainty ranges
print("\n=== Uncertainty Ranges (SSP3-7.0) ===")
ranges = proj.get_uncertainty_range('SSP3-7.0')
for year, low, high in ranges:
    print(f"{year}: [{low:.2f} - {high:.2f}]")
```

```python
# Plot with uncertainty
print(proj.render_uncertainty_fan('SSP3-7.0'))
```

```python
# 2050 values
print("\n=== PCSI in 2050 ===")
for scenario in ['SSP1-1.9', 'SSP3-7.0', 'SSP5-8.5']:
    data = proj.projections[scenario]
    for year, pcsi in data:
        if year == 2050:
            print(f"{scenario:10s}: {pcsi:.2f}")
```

```python
# Summary
print(proj.summary())
```
