# 5. Parameter Correlation Matrix
## Eight-Parameter Correlations 1960-2025

This notebook reproduces Figure 5 from the CARBONICA paper: Heatmap of correlations between all eight parameters.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica.visualization.correlation_matrix import CorrelationMatrix

# Initialize
corr = CorrelationMatrix()
```

```python
# Display correlation matrix
print(corr.render_text_matrix())
```

```python
# Get strongest correlations
print("\n=== Strongest Correlations ===")
strongest = corr.get_strongest_correlations(5)
for i, (p1, p2, r) in enumerate(strongest, 1):
    print(f"{i}. {p1:8s} ↔ {p2:8s}: {r:6.2f}")
```

```python
# Get positive feedbacks
print("\n=== Positive Feedbacks (r > 0.5) ===")
positive = corr.get_positive_feedbacks()
for p1, p2, r in positive:
    print(f"{p1:8s} ↔ {p2:8s}: {r:6.2f}")
```

```python
# Get negative feedbacks
print("\n=== Negative Feedbacks (r < -0.5) ===")
negative = corr.get_negative_feedbacks()
for p1, p2, r in negative:
    print(f"{p1:8s} ↔ {p2:8s}: {r:6.2f}")
```

```python
# Key correlations from paper
print("\n=== Key Correlations from Paper ===")
key_pairs = [
    ('G_atm', 'E_anth', 0.86, 'Direct emission link'),
    ('S_ocean', 'beta', -0.88, 'Ocean buffer chemistry'),
    ('F_perma', 'tau_soil', 0.77, 'Temperature driver'),
    ('NPP', 'Phi_q', 0.78, 'Photosynthetic efficiency')
]

for p1, p2, expected, desc in key_pairs:
    actual = corr.get_correlation(p1, p2)
    match = "✅" if abs(actual - expected) < 0.02 else "❌"
    print(f"{p1:8s} ↔ {p2:8s}: {actual:6.2f} (expected {expected:6.2f}) {match} - {desc}")
```

```python
# Export to CSV
csv_file = corr.to_csv("correlation_matrix.csv")
print(f"\n✅ Correlation matrix exported to {csv_file}")
```
