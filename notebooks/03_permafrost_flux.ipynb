# 3. Permafrost Thaw Flux
## Arctic Carbon Feedback 1960-2025

This notebook reproduces Figure 3 from the CARBONICA paper: Permafrost thaw flux reconstruction and projection.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica import CARBONICA
from carbonica.modules.permafrost_engine import PermafrostEngine
from carbonica.visualization.parameter_plots import ParameterPlots

# Initialize
carbonica = CARBONICA(data_dir="../data")
perma = PermafrostEngine()
plots = ParameterPlots(width=80)
```

```python
# Get historical flux
years = [1960, 1970, 1980, 1990, 2000, 2005, 2010, 2015, 2020, 2025]
flux_values = []

for year in years:
    flux = perma.get_flux(year)
    flux_values.append(flux)
    print(f"{year}: {flux:.2f} PgC/yr")
```

```python
# Calculate increase
print(f"\n=== Permafrost Flux Increase ===")
print(f"1990: {perma.get_flux(1990):.2f} PgC/yr")
print(f"2025: {perma.get_flux(2025):.2f} PgC/yr")
print(f"Increase: {perma.get_flux(2025)/perma.get_flux(1990):.0f}x (paper: 35x)")
```

```python
# Plot historical flux
print(plots.plot_timeseries(years, flux_values, "Permafrost Thaw Flux 1960-2025", "PgC/yr"))
```

```python
# Project future under different scenarios
print("\n=== Future Projections ===")
for scenario in ['SSP1-1.9', 'SSP3-7.0', 'SSP5-8.5']:
    proj = perma.project_flux(end_year=2050, scenario=scenario)
    print(f"\n{scenario}:")
    for year, flux in proj:
        print(f"  {year}: {flux:.2f} PgC/yr")
```

```python
# Abrupt vs gradual thaw
print("\n=== Abrupt vs Gradual Thaw (2025) ===")
components = perma.separate_thaw_types(1.71, 2025)
print(f"Total: {components['total']:.2f} PgC/yr")
print(f"Gradual: {components['gradual']:.2f} PgC/yr")
print(f"Abrupt: {components['abrupt']:.2f} PgC/yr")
print(f"Abrupt fraction: {components['abrupt_fraction']*100:.0f}% (paper: 31%)")
```

```python
# Regional breakdown
print("\n=== Regional Flux (2025) ===")
regions = ['western_siberia', 'canadian_zone', 'alaska', 'esas']
for region in regions:
    flux = perma.get_region_flux(region, 2025)
    print(f"{region:16s}: {flux:.2f} PgC/yr")
```

```python
# Critical thresholds
print("\n=== Critical Thresholds ===")
print(f"Self-sustaining threshold: 2.8 PgC/yr")
print(f"SSP3-7.0 critical year: {perma.get_critical_year('SSP3-7.0')} (paper: 2038-2046)")
print(f"ESAS risk window: {perma.get_esas_risk_year()[0]}-{perma.get_esas_risk_year()[1]}")
```

```python
# Global emissions percentage
total_emissions = carbonica.get_parameter('E_anth', 2025)
perma_flux = perma.get_flux(2025)
percentage = (perma_flux / total_emissions) * 100
print(f"\nPermafrost contribution to global emissions: {percentage:.1f}% (paper: 4.3%)")
```
