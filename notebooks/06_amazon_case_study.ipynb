# 6. Amazon Carbon Pivot Case Study
## From Sink to Source 2018-2023

This notebook reproduces Figure 6 from the CARBONICA paper: Analysis of the Amazon rainforest carbon balance transition.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica import CARBONICA
from carbonica.modules.quantum_yield_tracker import QuantumYieldTracker
from carbonica.visualization.parameter_plots import ParameterPlots

# Initialize
carbonica = CARBONICA(data_dir="../data")
qyt = QuantumYieldTracker()
plots = ParameterPlots(width=80)
```

```python
# Amazon Φ_q decline 2018-2023
years = [2018, 2019, 2020, 2021, 2022, 2023]
phi_amazon = []

for year in years:
    phi = qyt.get_quantum_yield(year, 'amazon_tropical')
    phi_amazon.append(phi)
    print(f"{year}: Φ_q = {phi:.4f}")
```

```python
# Calculate decline
phi_2018 = phi_amazon[0]
phi_2023 = phi_amazon[-1]
decline = (phi_2018 - phi_2023) / phi_2018 * 100
print(f"\nΦ_q decline 2018-2023: {decline:.1f}% (paper: 20%)")
```

```python
# Plot Φ_q decline
print(plots.plot_timeseries(years, phi_amazon, "Amazon Φ_q 2018-2023", "Φ_q"))
```

```python
# NPP reduction from Φ_q loss
npp_2018 = 0.8  # PgC/yr sink (from paper)
npp_loss = npp_2018 * (decline / 100)
print(f"\nNPP reduction from Φ_q loss: {npp_loss:.2f} PgC/yr")
```

```python
# Deforestation emissions
deforest_emissions = 0.5  # PgC/yr from paper
fire_emissions = 0.7  # PgC/yr from paper
print(f"\nDeforestation emissions: {deforest_emissions} PgC/yr")
print(f"Fire emissions: {fire_emissions} PgC/yr")
print(f"Total disturbance: {deforest_emissions + fire_emissions:.1f} PgC/yr")
```

```python
# Carbon balance reversal
sink_before = 0.8  # PgC/yr sink
source_after = 0.8  # PgC/yr source
reversal = sink_before + source_after
print(f"\n=== Carbon Balance Reversal ===")
print(f"Pre-2018 sink: {sink_before} PgC/yr")
print(f"2023 source: {source_after} PgC/yr")
print(f"Absolute deterioration: {reversal:.1f} PgC/yr (paper: 1.2 PgC/yr)")
```

```python
# Regional analysis
print("\n=== Southeastern Amazon ===")
print("First to reverse in 2021")
print("Area: 800,000 km²")
print("F_out = 1.2 PgC/yr, S_land = 0.4 PgC/yr → Net source 0.8 PgC/yr")
```

```python
# Soil carbon dynamics
print("\n=== Soil Carbon Effects ===")
print("Drought cycles cause:")
print("  • Slowed decomposition (short-term τ_soil increase)")
print("  • Reduced fresh carbon input")
print("  • Long-term soil carbon decline")
```

```python
# Future projection
print("\n=== Projection (if deforestation continues) ===")
print("2030: Amazon becomes net source of 1.5-2.0 PgC/yr")
print("Equivalent to adding entire EU fossil fuel emissions")
```
