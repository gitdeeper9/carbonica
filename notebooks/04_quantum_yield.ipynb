# 4. Photosynthetic Quantum Yield
## Φ_q Decline from Satellite SIF 2009-2025

This notebook reproduces Figure 4 from the CARBONICA paper: Global Φ_q trend from GOSAT and OCO-2 SIF retrievals.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica import CARBONICA
from carbonica.modules.quantum_yield_tracker import QuantumYieldTracker
from carbonica.visualization.parameter_plots import ParameterPlots

# Initialize
carbonica = CARBONICA(data_dir="../data")
qyt = QuantumYieldTracker()
plots = ParameterPlots(width=80)
```

```python
# Get global Φ_q time series
years = list(range(2009, 2026))
phi_global = []

for year in years:
    phi = qyt.get_quantum_yield(year, 'global_mean')
    phi_global.append(phi)
    print(f"{year}: Φ_q = {phi:.4f}")
```

```python
# Calculate trend
trend = qyt.get_trend(2009, 2025, 'global_mean')
print(f"\n=== Global Φ_q Trend ===")
print(f"2009: {trend['start_value']:.4f}")
print(f"2025: {trend['end_value']:.4f}")
print(f"Absolute change: {trend['absolute_change']:.4f}")
print(f"Relative change: {trend['relative_change']:.2f}%")
print(f"Decline rate: {trend['decline_rate_per_decade']:.2f}%/decade (paper: -0.9%/decade)")
```

```python
# Plot global Φ_q
print(plots.plot_timeseries(years, phi_global, "Global Φ_q 2009-2025", "Φ_q"))
```

```python
# Biome-specific analysis
print("\n=== Biome-specific Φ_q (2025) ===")
biomes = ['amazon_tropical', 'se_asian_peat', 'mediterranean', 'boreal_forest', 'global_mean']
for biome in biomes:
    phi = qyt.get_quantum_yield(2025, biome)
    print(f"{biome:20s}: {phi:.4f}")
```

```python
# Biome decline rates
print("\n=== Biome Decline Rates (%/decade) ===")
for biome, rate in qyt.decline_rates.items():
    print(f"{biome:20s}: {rate:+.1f}%")
```

```python
# Stress detection
print("\n=== Stress Detection (2025) ===")
for biome in ['amazon_tropical', 'boreal_forest']:
    phi = qyt.get_quantum_yield(2025, biome)
    stressed, level = qyt.detect_stress(phi, biome)
    print(f"{biome:20s}: Φ_q = {phi:.4f} → {level}")
```

```python
# Compare with MODIS NPP
npp_2009 = carbonica.budget.get_npp(2009)
npp_2025 = carbonica.budget.get_npp(2025)
print(f"\n=== NPP Change ===")
print(f"NPP 2009: {npp_2009:.1f} PgC/yr")
print(f"NPP 2025: {npp_2025:.1f} PgC/yr")
print(f"NPP change: {npp_2025 - npp_2009:.1f} PgC/yr")
```

```python
# SIF data (simulated)
print("\n=== GOSAT SIF Data (Amazon) ===")
gosat_data = qyt.load_gosat()
for record in gosat_data[:5]:  # First 5 records
    print(f"{record['year']}: SIF = {record['sif_740']} mW/m²/sr/nm")
```
