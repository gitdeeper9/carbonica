# 7. Arctic Carbon System Case Study
## Planetary Tipping Edge

This notebook reproduces Figure 7 from the CARBONICA paper: Analysis of Arctic permafrost carbon feedback.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica import CARBONICA
from carbonica.modules.permafrost_engine import PermafrostEngine
from carbonica.visualization.parameter_plots import ParameterPlots

# Initialize
carbonica = CARBONICA(data_dir="../data")
perma = PermafrostEngine()
plots = ParameterPlots(width=80)
```

```python
# Arctic warming
print("=== Arctic Warming ===")
warming_arctic = 3.8  # °C since 1850
warming_global = 1.2  # °C since 1850
amplification = warming_arctic / warming_global
print(f"Arctic warming: +{warming_arctic}°C")
print(f"Global warming: +{warming_global}°C")
print(f"Arctic amplification: {amplification:.1f}x (paper: 3.2x)")
```

```python
# Carbon stocks
print("\n=== Arctic Carbon Stocks ===")
stocks = perma.get_carbon_stocks()
print(f"Total permafrost carbon: {stocks['global']['carbon_pg']} PgC")
print(f"Northern peatlands: ~500 PgC")
print(f"ESAS submarine: unknown but potentially large")
```

```python
# GTN-P borehole network
print("\n=== GTN-P Monitoring Network ===")
print(f"Boreholes: {perma.global_totals['n_boreholes']} stations")
print("Regions: Siberia, Alaska, Canada, Scandinavia")
```

```python
# Regional flux distribution
print("\n=== Regional Flux Distribution (2025) ===")
total_2025 = perma.get_flux(2025)
regions = ['western_siberia', 'canadian_zone', 'alaska', 'esas']

for region in regions:
    flux = perma.get_region_flux(region, 2025)
    pct = (flux / total_2025) * 100
    print(f"{region:16s}: {flux:.2f} PgC/yr ({pct:.0f}%)")
```

```python
# Western Siberia + Canadian zone dominate
ws_flux = perma.get_region_flux('western_siberia', 2025)
cz_flux = perma.get_region_flux('canadian_zone', 2025)
dominant = (ws_flux + cz_flux) / total_2025 * 100
print(f"\nWestern Siberia + Canadian zone: {dominant:.0f}% of total (paper: 73%)")
```

```python
# ESAS submarine permafrost
print("\n=== East Siberian Arctic Shelf (ESAS) ===")
esas_data = perma.regional_data['esas']
print(f"Area: {esas_data['area_km2']/1e6:.1f} million km²")
print(f"Carbon density: {esas_data['carbon_density']} kgC/m³ (5× terrestrial)")
print(f"MAGT 2025: {esas_data['magt_2025']}°C")
print(f"Warming trend: {esas_data['magt_trend']}°C/yr")
print(f"Risk window: {perma.get_esas_risk_year()[0]}-{perma.get_esas_risk_year()[1]}")
```

```python
# CH₄ vs CO₂
print("\n=== Methane Considerations ===")
print("ESAS releases CH₄ under anaerobic conditions")
print("CH₄ GWP: 84× CO₂ over 20 years")
print("Atmospheric lifetime: 12 years")
```

```python
# Projections
print("\n=== Future Projections ===")
proj = perma.project_flux(end_year=2060, scenario='SSP3-7.0')
for year, flux in proj:
    if year in [2040, 2060]:
        print(f"{year}: {flux:.1f} PgC/yr")
```

```python
# Comparison with EU emissions
eu_emissions = 0.9  # PgC/yr
print(f"\nEU emissions: {eu_emissions} PgC/yr")
print(f"2040 permafrost flux: equivalent to {proj[3][1]/eu_emissions:.1f}× EU")
```
