# 9. Structural Equation Modeling
## Carbon Cycle Causal Pathways

This notebook reproduces Figure 9 from the CARBONICA paper: SEM analysis of carbon cycle feedback pathways.

```python
# Import CARBONICA
import sys
sys.path.append('..')
from carbonica.statistics.sem import StructuralEquationModel

# Initialize
sem = StructuralEquationModel()
```

```python
# Display model structure
print("=== SEM Causal Pathways ===\n")
print("Direct pathways:")
for a, b in sem.pathways['direct']:
    coef = sem.get_path_coefficient(a, b)
    print(f"  {a:8s} → {b:8s} : {coef:6.2f}")
```

```python
# Get strongest pathways
print("\n=== Strongest Pathways ===")
strongest = sem.get_strongest_pathways(5)
for i, (a, b, coef) in enumerate(strongest, 1):
    print(f"{i}. {a:8s} → {b:8s} : {coef:6.2f}")
```

```python
# Calculate total effects
print("\n=== Total Effects ===")
effects = [
    ('E_anth', 'G_atm'),
    ('G_atm', 'beta'),
    ('G_atm', 'F_perma'),
    ('F_perma', 'G_atm'),
    ('Phi_q', 'NPP'),
    ('NPP', 'S_land')
]

for cause, effect in effects:
    total = sem.get_total_effect(cause, effect)
    direct = sem.get_path_coefficient(cause, effect)
    print(f"{cause:8s} → {effect:8s}: total = {total:6.2f}, direct = {direct:6.2f}")
```

```python
# Fit statistics
print("\n=== Fit Statistics ===")
stats = sem.fit_stats
print(f"CFI   : {stats['CFI']:.3f} (≥0.95 excellent)")
print(f"RMSEA : {stats['RMSEA']:.3f} (≤0.05 excellent)")
print(f"SRMR  : {stats['SRMR']:.3f}")
print(f"χ²/df : {stats['chi2']/stats['df']:.2f}")
```

```python
# Implied correlations
print("\n=== Implied Correlations ===")
variables = ['G_atm', 'E_anth', 'beta', 'S_ocean']
for i, v1 in enumerate(variables):
    for v2 in variables[i+1:]:
        corr = sem.compute_correlation(v1, v2)
        print(f"{v1:8s} ↔ {v2:8s} : {corr:6.2f}")
```

```python
# Compare with observed correlations
print("\n=== Model vs Observed ===")
from carbonica.visualization.correlation_matrix import CorrelationMatrix
corr_obs = CorrelationMatrix()

pairs = [('G_atm', 'E_anth'), ('G_atm', 'beta'), ('S_ocean', 'beta')]
for v1, v2 in pairs:
    model_corr = sem.compute_correlation(v1, v2)
    obs_corr = corr_obs.get_correlation(v1, v2)
    diff = abs(model_corr - obs_corr)
    print(f"{v1:8s}-{v2:8s}: model={model_corr:6.2f}, obs={obs_corr:6.2f}, diff={diff:.3f}")
```

```python
# Key findings from paper
print("\n=== Key Findings ===")
print("• G_atm → β : -0.91 (strongest negative)")
print("• NPP → S_land : 0.94 (strongest positive)")
print("• E_anth → G_atm : 0.86 (direct link)")
print("• F_perma → G_atm : 0.67 (positive feedback)")
```

```python
# Summary
print(sem.summary())
```
